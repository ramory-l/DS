```python
import numpy as np
import matplotlib.pyplot as plt
import random
import statsmodels.api as sm
import matplotlib.pyplot as plt

%matplotlib inline
```

Сгенерируйте $100$ значений временного ряда по правилу$X_t=0.2X_{t-1}+0.1X_{t-2}+ \varepsilon_t$  
В качестве шумовой компоненты возьмите нормальный белый шум с нулевым средним и единичной дисперсией.

```python
# num - число шагов
# a, b - множители. В каждый момент времени следующее значение равно a*предыдущее + b*предпредыдущее + белый шум
# mu, sigma - параметры шума
def normrwalk(num, a, b, mu, sigma):
    eps = np.random.normal(mu, sigma, size=num)
    walk = np.zeros(num)
    walk[0] = eps[0]
    for i in range (2,num):
        walk[i] = a*walk[i-1] + b*walk[i-2] + eps[i]
    return walk
```

Постройте его график.

```python
a = 0.2
b = 0.1
n_walk_02_01 = normrwalk(100, a, b, 0, 1)
plt.plot(n_walk_02_01)
```

Проверьте его на стационарность по критерию Дикки-Фуллера:

```python
test = sm.tsa.adfuller(n_walk_02_01)
print ('p-value: ', test[1])
if test[1]< 0.05: 
    print (' ряд стационарен')
else:
    print ('ряд не стационарен')
```

```python
%pylab inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/AirPassengers.csv')
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Month'], format = '%Y-%m')
```

```python
df = df.set_index(pd.DatetimeIndex(df['Month']))
df.drop(['Month'], axis =1, inplace = True)
```

```python
df.head()
```

```python
passengers = df['#Passengers']
```

```python
plt.figure(figsize(17,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(passengers.squeeze(), lags=143, ax=ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(passengers.squeeze(), lags=71, ax=ax)
plt.show()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(passengers)[1])
```

Первая разность

```python
passengers = df['#Passengers'].diff().dropna()
```

```python
plt.figure(figsize(17,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(passengers.squeeze(), lags=142, ax=ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(passengers.squeeze(), lags=70, ax=ax)
plt.show()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(passengers)[1])
```

```python
plt.plot(passengers)
```

Вторая разность

```python
passengers = df['#Passengers'].diff(periods=2).dropna()
```

```python
plt.figure(figsize(17,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(passengers.squeeze(), lags=140, ax=ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(passengers.squeeze(), lags=69, ax=ax)
plt.show()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(passengers)[1])
```
